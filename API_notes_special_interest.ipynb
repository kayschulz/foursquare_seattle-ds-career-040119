```python
import json
import requests
```

```python
# Load secrets from credentials.json
url = 'https://api.foursquare.com/v2/venues/explore'
with open('.secrets/credentials.json') as f:
    params = json.load(f)
```

```python
params['v'] = '20180323' #version, bc apis may not be consistent. So you can use the old version at later date
params['ll'] = '47.608,-122.336', #lat-long
params['query'] = 'thai',
params['limit'] = 100 #how many max to give back

response = requests.get(url=url, params=params)
data = json.loads(response.text)
```

```python
# Goal to find how many thai restaurants deliver
len(data['response']['groups'][0]['items'])
```

```python
results = data['response']['groups'][0]['items']
results[0]
```

```python
import pandas as pd
pd.DataFrame(results)
```

```python
venue_names = []
for result in results:
    if 'delivery' in result['venue'].keys():
        venue_names.append(result['venue']['name'])
```
