```python
import json
import requests
```

```python
# Load secrets from credentials.json
url = 'https://api.foursquare.com/v2/venues/explore'
with open('.secrets/credentials.json') as f:
    params = json.load(f)
```

```python
params['v'] = '20180323' #version, bc apis may not be consistent. So you can use the old version at later date
params['ll'] = '47.608,-122.336', #lat-long
params['query'] = 'pilates',
params['limit'] = 100 #how many max to give back

response = requests.get(url=url, params=params)
data = json.loads(response.text)
```

```python
data.keys()
```

```python
data['response'].keys()
```

```python
# Goal to find how many pilates studios around flatiron
# we add the zero because all results were returned as a single item list
len(data['response']['groups'][0]['items'])
```

```python
results = data['response']['groups'][0]['items']
results
```

```python
import pandas as pd
pd.DataFrame(results)
# still to vauge as what we want is further in the venue
```

```python
venues = []
for result in results:
    venues.append(result['venue'])
```

```python
pd.DataFrame(venues).head().T
```

```python
# now to store in mongodb
import pymongo
```

```python
mc = pymongo.MongoClient()
db = mc['foursquare']
pilates_coll = db['pilates']

pilates_coll.insert_many(venues)
```
